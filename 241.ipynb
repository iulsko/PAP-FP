## Wechselstromeigenschaften von RLC-Gliedern

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
```

```python
#rg = 680kOhm
plt.figure(1)
plt.title("Frequenzgang des Verstärkers beim Widerstand von 680 k$\Omega$")
plt.ylabel('Verstärkung []')
plt.xlabel('Frequenz [Hz]')
plt.xscale('log')
plt.yscale('log')
x_old = np.array([0.1,0.4,0.75,1.0,5.0,8.0,20.0,70.0,90.0,200,600,900])
x = [num*1000 for num in x_old]
y_old = np.array([2.84,2.83,2.75,2.71,1.45,1,0.411,0.121,0.0926,0.042,0.0137,0.0095])
y = [num/0.3 for num in y_old]
yerr_old = np.array([40,10,10,10,10,10,4,2,1,1,2,1])
yerr = [num/300 for num in yerr_old]
plt.errorbar(x, y, yerr=yerr, marker='o', markersize=4, linestyle='none', color = 'g', label = 'Messwerte')
plt.legend(loc = 3)
plt.grid(True)

plt.show()
```
