## Energiekalebrierung

```python
kanal=np.array([842, 943, 494, 617])
kanal_fehler = np.array([30, 20, 24, 27])
energie=np.array([1.17323, 1.33248, 0.6616, 0.8353])


def linear(x,c):
    return c*x

popt, pcov = curve_fit(linear, kanal, energie, sigma = kanal_fehler)
print('Steigung = ', popt[0], '[MeV/Kanal]')
print('Fehler der Steigung = ', np.sqrt(pcov[0][0]), '[MeV/Kanal]')

plt.figure(figsize=(15,10))
l = np.linspace(400,1000,10000)
plt.errorbar(kanal, energie, xerr=kanal_fehler, fmt='.', label = 'Messwerte', color = 'darkcyan')
plt.plot(l, linear(l, *popt), label = 'Fitgerade', color = 'goldenrod')
plt.grid()
plt.xlabel('Kanalnummer', fontsize = 13)
plt.ylabel('Energie [MeV]', fontsize = 13)
plt.legend(loc = 4)
plt.title('Energiekalibrierung fr Co$^{60}$, Cs$^{137}$ & Mn$^{54}$', fontsize = 18)
#plt.savefig('./energiekal1.png')
plt.show()
```

```python
kanal=np.array([395, 933, 38])
kanal_fehler = np.array([20, 40, 5])
energie=np.array([0.511, 1.2746, 0.081])


def linear(x,c):
    return c*x

popt, pcov = curve_fit(linear, kanal, energie, sigma = kanal_fehler)
print('Steigung = ', popt[0], '[MeV/Kanal]')
print('Fehler der Steigung = ', np.sqrt(pcov[0][0]), '[MeV/Kanal]')

plt.figure(figsize=(15,10))
l = np.linspace(50,1000,10000)
plt.errorbar(kanal, energie, xerr=kanal_fehler, fmt='.', label = 'Messwerte', color = 'darkcyan')
plt.plot(l, linear(l, *popt), label = 'Fitgerade', color = 'goldenrod')
plt.grid()
plt.xlabel('Kanalnummer', fontsize = 13)
plt.ylabel('Energie [MeV]', fontsize = 13)
plt.legend(loc = 4)
plt.title('Energiekalibrierung fr Ba$^{133}$ & Na$^{22}$', fontsize = 18)
#plt.savefig('./energiekal2.png')
plt.show()
```
