## Spannungsverstärkung

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
for i in range(0,1):
    ein = input("eingang in mv + fehler")
    aus = input("ausgang in v + fehler")

    liste = ein.split(',')
    e = float(liste[0]) / 1000
    efehler = float(liste[1]) /1000

    lista = aus.split(',')
    a = float(lista[0])
    afehler = float(lista[1])

    v = -a/e
    vfehler = sqrt((afehler/e)**2 + (a*efehler/e/e)**2)
    print(v)
    print(vfehler)
```

```python
plt.figure(1)

x = np.array([84.8,70.4,50.8,37.2,17,4.3]) #mV
y = np.array([7.4,6.28,4.58,3.34,1.48,0.37]) #V
xerr = np.array([0.5,0.5,0.4,0.1,0.2,0.1]) #mV
yerr = np.array([0.1,0.05,0.03,0.02,0.01,0.002]) #V

a = np.array([4.3,84.8])
a_new = [num/1000 for num in a]
b = np.array([0.372,7.3])

#ausgleichsgerade
fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit)

x_new = [num/1000 for num in x]
fit_new = np.polyfit(x_new,y,1)
print("Steigung Ausgleichsgerade:")
print(fit_new)

slope, intercept = np.polyfit(a_new, b, 1)
print("Steigung Fehlergerade:")
print(slope, intercept)

# fit_fn is now a function which takes in x and returns an estimate for y
plt.plot(x, fit_fn(x), color='g', label = 'Widerstand')
plt.plot(a, b, color = 'r', label = 'Fehlergerade')
plt.errorbar(x, y, xerr=xerr, yerr=yerr, marker='o', markersize=2, linestyle='none', color = 'g')
plt.legend(loc=2,ncol=1)
plt.grid(True)
plt.title('Wechselspannungsverstärkung mit dem Widerstand von 274 k${\Omega}$ ')

plt.figure(2)

plt.ylabel('Ausgangsspannung [V]')
plt.xlabel('Eingangsspannung [mV]')

x = np.array([4.2,12.4,21.2,41.1,62.2,80.0])
y = np.array([0.888,2.74,4.45,8.88,13.5,17.3])

#wir ignore the first and the last point
xerr = np.array([0.1,0.1,0.2,0.1,0.2,0.2])
yerr = np.array([0.001,0.01,0.01,0.01,0.1,0.1])

a = np.array([4.2,80.0])
a_new = [num/1000 for num in a]
b = np.array([0.889,17.2])

fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit)

x_new = [num/1000 for num in x]
fit_new = np.polyfit(x_new,y,1)
print("Steigung Ausgleichsgerade:")
print(fit_new)

slope, intercept = np.polyfit(a_new, b, 1)
print("Steigung Fehlergerade:")
print(slope, intercept)

plt.plot(x, fit_fn(x), color='g', label = 'Widerstand')
plt.plot(a, b, color = 'r', label = 'Fehlergerade')
plt.errorbar(x, y, xerr=xerr, yerr=yerr, marker='o', markersize=2, linestyle='none', color='g')
plt.legend(loc=2,ncol=1)
plt.grid(True)
plt.title('Wechselspannungsverstärkung mit dem Widerstand von 680 k${\Omega}$ ')

# Diagramm anzeigen:
plt.show()
```
