## Mehanik und Vakuum

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
data = pd.read_excel('C:\\Users\\user\\Desktop\\Messung des Saugvem__gens.xlsx')
druck = data['Druck p [mbar]']
fehler_druck = data['Fehler Druck [mbar]']
saug = data['Saugvermoegen [l/s]']
fehler_saug = data['Fehler Saugvermoegen [l/s]']
sauglei = data['Saugleistung[l*mbar/s]']
fehler_sauglei = data['Saugleistung Fehler [l*mbar/s]']
d = []
fd = []
s = []
fs = []
sl = []
fsl = []

for i in range(0,len(druck)):
    d.append(druck[i]) 
for i in range(0,len(fehler_druck)):
    fd.append(fehler_druck[i])
for i in range(0,len(saug)):
    s.append(saug[i])
for i in range(0,len(fehler_saug)):
    fs.append(fehler_saug[i])
for i in range(0,len(fehler_sauglei)):
    fsl.append(fehler_sauglei[i])
for i in range(0,len(sauglei)):
    sl.append(sauglei[i])

d = np.asarray(d)
fd = np.asarray(fd)
s = np.asarray(s)
fs = np.asarray(fs)
sl = np.asarray(sl)
fsl = np.asarray(fsl)

plt.figure(1)
plt.errorbar(d, s, yerr=fs, xerr=fd, fmt='.', label = 'Messwerte', capsize = 2)
plt.title('Diagramm 1')
plt.xlabel('Druck [mbar]')
plt.ylabel('Saugvermoegen [l/s]')
plt.legend(loc = 'best')
plt.grid(True,which="both")
plt.xscale('log')

plt.figure(2)
plt.errorbar(d, sl, yerr=fsl, xerr=fd, fmt='.', label = 'Messwerte', capsize = 2)
plt.title('Diagramm 2')
plt.xlabel('Druck [mbar]')
plt.ylabel('Saugleistung [l*mbar/s]')
plt.legend(loc = 'best')
plt.grid(True,which="both")
plt.yscale('log')
plt.xscale('log')
plt.show()
```
